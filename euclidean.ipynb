```python
np.random.seed(71)
result = []
for num in range(0,1000000):
  n_data = 128
  a = np.random.randint(-127, 128, size=n_data)
  b = np.random.randint(-127, 128, size=n_data)
  c = np.linalg.norm(a-b)
  result.append(c)
```

```python
plt.figure(figsize=(8,4))
plt.hist(result, bins=np.arange(900,1500,10),)
plt.show
```
